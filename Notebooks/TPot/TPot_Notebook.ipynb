```python
# importing useful packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tpot import TPOTClassifier
from tpot import TPOTRegressor

## importing the data
data = pd.read_csv('../Pipelines/MinMax1_Final.csv', index_col=0, header=0)
data.head()
```

```python
data = data.astype(int)
data.info()
```

```python
# shuffling data
data_shuffle = data.iloc[np.random.permutation(len(data))]
data_reset = data_shuffle.reset_index(drop=True)
data_reset.head()
```

```python
# taking our target feature `Y`
data_Y = data_reset['Y'].values
```

```python
# splitting data
training_indicies, testing_indicies = train_test_split(data_reset.index, stratify = data_Y, train_size=0.75, test_size=0.25, random_state=2019)
training_indicies.size, testing_indicies.size
```

```python
# Using TPot now
tpot_good = TPOTClassifier(verbosity=2, max_time_mins=2, max_eval_time_mins=0.04, population_size=20, generations=5, scoring='roc_auc')
tpot_good.fit(data_reset.drop('Y', axis=1).loc[training_indicies].values, data_reset.loc[training_indicies,'Y'].values)
```

```python
tpot_good.export('tpot_good_model.py')
```
